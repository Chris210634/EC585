```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("./christopherliao.jpg")
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB) )
plt.title("Original Image")
plt.show()
```

```python
# I will convert to greyscale by taking average of R, G, and B channels
height = image.shape[0]
width = image.shape[1]
for i in range(height):
    for j in range(width):
        average = (int(image[i,j,1]) + int(image[i,j,2]) + int(image[i,j,0])) / 3
        image[i,j,0] = average
        image[i,j,1] = average
        image[i,j,2] = average
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB) )
plt.title("Grayscale Image")
plt.show()
```

```python
cv2.imwrite("grayscale.jpg", image)
```

```python
# Let's try applying a horizontal edge filter [-1, 0, +1]
filtered_image = image.copy()
height = image.shape[0]
width = image.shape[1]
for i in range(height):
    for j in range(width):
        if j == 0:
            output = 0
        elif j == width - 1:
            output = 0
        else:
            output = abs(int(image[i,j+1,1]) - int(image[i,j-1,1]))*2
        filtered_image[i,j,0] = output
        filtered_image[i,j,1] = output
        filtered_image[i,j,2] = output
```

```python
plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB) )
plt.title("Horizontal-edge Filtered Image")
plt.show()
```

```python
cv2.imwrite("filtered.jpg", filtered_image)
```
